```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
```

```python
# Utiliser l'affichage total
pd.set_option('display.max_columns',None)
```

```python
Train_set=pd.read_csv('train.csv')
Test_set=pd.read_csv('test.csv')
Train_set.head()
print(Train_set)
```

```python
Train_set_mask= Train_set.dtypes=="object"
categorical_cols=list(Train_set.columns[Train_set_mask])
ordinal_encoder_train = OrdinalEncoder()
Encoded_train_set=pd.DataFrame(ordinal_encoder_train.fit_transform(Train_set,categorical_cols))

simple_imputer_train=SimpleImputer()
Encoded_train_set=pd.DataFrame(simple_imputer_train.fit_transform(Encoded_train_set))
```

```python
sns.heatmap(Encoded_train_set.corr())
Encoded_train_set.corr()
```

```python
Test_set_mask= Test_set.dtypes=="object"
categorical_cols=list(Test_set.columns[Test_set_mask])
ordinal_encoder_test = OrdinalEncoder()
Encoded_Test_set=pd.DataFrame(ordinal_encoder_test.fit_transform(Test_set,categorical_cols))

simple_imputer_test=SimpleImputer()
Encoded_Test_set=pd.DataFrame(simple_imputer_test.fit_transform(Encoded_Test_set))
```

```python
sns.heatmap(Encoded_Test_set.corr())
Encoded_Test_set.corr()
```
